Download the 'owid-covid-data.csv' dataset from Our World in Data and save it in the working folder.

# Data Loading & Exploration
Use pandas to load the dataset, check its structure with `df.columns` and `df.head()`, and identify missing values using `df.isnull().sum()`.

```python
# Data Loading & Exploration
import pandas as pd

# Load the dataset
file_path = 'owid-covid-data.csv'
df = pd.read_csv(file_path)

# Check the structure of the dataset
print("Columns in the dataset:")
print(df.columns)

# Preview the first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Identify missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

# Data Cleaning
Filter countries of interest, drop rows with missing critical values, convert the date column to datetime, and handle missing numeric values using `fillna()` or `interpolate()`.

```python
# Data Cleaning
# Filter countries of interest
countries_of_interest = ['Kenya', 'USA', 'India']
filtered_df = df[df['location'].isin(countries_of_interest)]

# Drop rows with missing critical values
filtered_df = filtered_df.dropna(subset=['date', 'total_cases', 'total_deaths'])

# Convert the date column to datetime
filtered_df['date'] = pd.to_datetime(filtered_df['date'])

# Handle missing numeric values
filtered_df = filtered_df.fillna(method='ffill')  # Forward fill as an example
```

# Exploratory Data Analysis (EDA)
Generate descriptive statistics and visualize trends such as total cases and deaths over time, daily new cases, and death rates using matplotlib and seaborn.

```python
# Exploratory Data Analysis (EDA)
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total cases over time for selected countries
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)
plt.title('Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.show()

# Plot total deaths over time
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)
plt.title('Total Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.show()

# Compare daily new cases between countries
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)
plt.title('Daily New Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

# Visualizing Vaccination Progress
Plot cumulative vaccinations over time and compare the percentage of vaccinated populations using line and pie charts.

```python
# Visualizing Vaccination Progress
# Plot cumulative vaccinations over time
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)
plt.title('Cumulative Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.show()
```

# Optional: Build a Choropleth Map
Use Plotly Express or geopandas to create a choropleth map showing case density or vaccination rates by country.

```python
# Optional: Build a Choropleth Map
import plotly.express as px

# Prepare data for the latest date
latest_date = filtered_df['date'].max()
choropleth_data = filtered_df[filtered_df['date'] == latest_date]

# Create a choropleth map
fig = px.choropleth(
    choropleth_data,
    locations='iso_code',
    color='total_cases',
    hover_name='location',
    title='COVID-19 Case Density by Country',
    color_continuous_scale=px.colors.sequential.Plasma
)
fig.show()
```
